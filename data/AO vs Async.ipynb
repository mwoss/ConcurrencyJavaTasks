```python
import pandas as pd
```

```python
ao10 = pd.read_csv('AO/outputAO10.txt', sep=" ", header=None, names=["time", "cons", "prods"])
async10 = pd.read_csv('Async/outputAsync10.txt', sep=" ", header=None, names=["time", "cons", "prods"])
```

```python
time = ao10['time'].values
ao_cons = ao10['cons'].values
async_cons = async10['cons'].values
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(16,10))
plt.plot(time, ao_cons, 'ro')
plt.plot(time, async_cons, 'ro', color='blue')
plt.show()
```
