```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
```

```python
ao10 = pd.read_csv('AO/new_outputAO100.txt', sep=" ", header=None, names=["time", "cons", "prods"])
async10 = pd.read_csv('Async/new_outputAsync100.txt', sep=" ", header=None, names=["time", "cons", "prods"])
```

```python
time = ao10['time'].values
ao_cons = ao10['cons'].values
async_cons = async10['cons'].values
```

```python
plt.figure(figsize=(16,10))
plt.title('100 threads | Random task data | Bufor size = 1000 | Small overload on task = 25ms')
plt.xlabel('Czas')
plt.ylabel('Ilosc wykonnanych taskow')
red_patch = mpatches.Patch(color='red', label='AO')
blue_patch = mpatches.Patch(color='blue', label='Async')
plt.legend(handles=[red_patch, blue_patch])
plt.plot(time, ao_cons, 'ro')
plt.plot(time, async_cons, 'ro', color='blue')
plt.show()
```
